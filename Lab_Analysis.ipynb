## PH level of water

The ideal ph level for irrigatation water in greenhouse is between 5.2 to 6.0.So filtering out the values that do not fall in this range.

```python
df[(df['irr_PH'] <= 6.0) & (df['irr_PH'] >= 5.2)]
```

```python
df[(df['irr_EC'] < 4.0) & (df['irr_EC'] >= 1.0)]
```
